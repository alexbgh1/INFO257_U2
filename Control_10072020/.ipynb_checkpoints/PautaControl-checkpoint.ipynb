**1.** Describa en qué consiste el análisis de componentes principales y cuáles son sus aplicaciones más comunes

El análisis de componentes principales (PCA) es el proceso mediante el cual se define una proyección de los datos originales multidimensionales en un espacio de dimensión menor, habitualmente 2, el llamado plano principal. Para ello, mediante propiedades geométricas, se calcula la proyección que maximiza la varianza empirica de los datos en direcciones ortogonales. 
El PCA es un tipo de aprendizaje supervisado y sus principales aplicaciones son 
i) el análisis exploratorio de datos, como una primera herramienta de visualización de datos para reducir el número de variables y evitar la multicolinealidad, o cuando se tienen demasiados predictores en relación con el número de observaciones.
ii) en la reducción de la dimensionalidad de los datos
iii) en tareas de visualización de otros análisis, como por ejemplo clustering. 


**2.** Describa uno de los algoritmos de agrupamiento que hemos estudiado, indique ventajas, desventajas y dé un ejemplo de aplicación.



GMM es un algoritmo de agrupamiento que supone que los datos provienen de una mezcla de k normales multivariadas cuya estructura no es observable, es decir cada observación proviene de alguna de las k distribuciones, pero no está disponible la información de cual de ellas es.
Para su estimación se utiliza el algoritmo EM.
Como desventaja se puede considerar que a diferencia de otros métodos requiere que se cumpla la hipótesis de normalidad de las observaciones, pero al mismo tiempo este enfoque da mayor flexibilidad el análisis, pues cada observación no es asignada de manera determinística a un grupo si no que en probabilidad. Esto permite particularmente modelar adecuadamente el caso en que hay traslape entre grupos.  También es útil cuando la estructura de varianzas-covarianzas es distinta en cada grupo. Otra posible desventaja es que su estimación (algoritmo EM) puede ser costosa en términos de cálculo computacional.
Un ejemplo clásico de aplicación es el caso de datos que consideran ruido tales como datos de fallas geológicas, clima, astronomía, por nombrar algunos tipos.
